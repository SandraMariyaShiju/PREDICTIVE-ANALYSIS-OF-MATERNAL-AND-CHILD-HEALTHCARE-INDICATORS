```python
import tkinter as tk
from tkinter import ttk
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from statsmodels.tsa.arima.model import ARIMA

# Load your dataset
df = pd.read_csv("model.csv")

# List of specific state names
specific_states = [
    'Jammu & Kashmir', 'Himachal Pradesh', 'Meghalaya', 'Sikkim', 'Uttar Pradesh',
    'Arunachal Pradesh', 'Bihar', 'Nagaland', 'West Bengal', 'Andhra Pradesh',
    'A & N Islands', 'Assam', 'Mizoram', 'Chhattisgarh', 'Chandigarh',
    'Dadra & Nagar Haveli', 'Daman & Diu', 'Delhi', 'Goa', 'Gujarat',
    'Haryana', 'Jharkhand', 'Madhya Pradesh', 'Karnataka', 'Manipur',
    'Kerala', 'Lakshadweep', 'Maharashtra', 'Odisha', 'Puducherry',
    'Punjab', 'Rajasthan', 'Tripura', 'Telangana', 'Tamil Nadu', 'Uttarakhand'
]

def calculate_forecast(selected_state, selected_year, selected_variable):
    # Subset the data for the selected state and variable
    state_data = df[df['State'] == selected_state].groupby('YearCode')[selected_variable].sum()

    # Train the ARIMA model for the selected state and variable
    model = ARIMA(state_data, order=(1, 1, 1))  # ARIMA with (p, d, q) = (1, 1, 1)
    arima_results = model.fit()

    # Forecast the variable for the selected year
    forecast_steps = selected_year - max(state_data.index)
    forecast = arima_results.forecast(steps=forecast_steps)
    forecasted_values = [round(value) for value in forecast]

    return forecasted_values

def plot_forecast_results(actual_data, forecasted_values, selected_state, selected_year, selected_variable, frame):
    # Create a new figure
    fig = plt.Figure(figsize=(6, 4))
    ax = fig.add_subplot(111)

    # Plot actual data
    ax.plot(actual_data.index, actual_data.values, marker='o', linestyle='-', color='blue', label=f'Actual {selected_variable}')

    # Plot forecasted data
    forecast_years = list(range(max(actual_data.index) + 1, selected_year + 1))
    ax.plot(forecast_years, forecasted_values, marker='x', linestyle='-', color='red', label=f'Forecasted {selected_variable}')

    ax.set_title(f'{selected_variable.capitalize()} Forecast for {selected_state} in {selected_year} (ARIMA)')
    ax.set_xlabel('Year')
    ax.set_ylabel(f'{selected_variable.capitalize()}')
    ax.grid(True)
    ax.legend()

    # Create canvas to display the plot
    canvas = FigureCanvasTkAgg(fig, master=frame)
    canvas.draw()
    canvas.get_tk_widget().pack()

# Create the main window
root = tk.Tk()
root.title("Maternal Deaths and Infant Deaths Forecaster") 

# Increase the size of the window
root.geometry("800x600")

# Create a frame to contain the plot
frame = tk.Frame(root)
frame.pack()

def on_button_click():
    # Clear previous plot
    for widget in frame.winfo_children():
        widget.destroy()

    selected_state = state_var.get()
    selected_year = int(year_var.get())
    selected_variable = variable_var.get()

    # Calculate and plot forecast for the selected state, year, and variable
    actual_data = df[df['State'] == selected_state].groupby('YearCode')[selected_variable].sum()
    forecasted_values = calculate_forecast(selected_state, selected_year, selected_variable)
    plot_forecast_results(actual_data, forecasted_values, selected_state, selected_year, selected_variable, frame)

# State selection
state_label = tk.Label(root, text="Select a State:")
state_label.pack()
state_var = tk.StringVar()
state_dropdown = ttk.Combobox(root, textvariable=state_var, values=specific_states)
state_dropdown.pack()

# Year selection
year_label = tk.Label(root, text="Select a Year:")
year_label.pack()
year_var = tk.StringVar()
year_dropdown = ttk.Combobox(root, textvariable=year_var, values=['2020', '2021'])
year_dropdown.pack()

# Variable selection (Maternal Deaths or Infant Deaths Reported)
variable_label = tk.Label(root, text="Select a Variable:")
variable_label.pack()
variable_var = tk.StringVar()
variable_dropdown = ttk.Combobox(root, textvariable=variable_var, values=['Maternal_Deaths', 'Infant deaths reported'])
variable_dropdown.pack()

# Button to plot forecast
plot_button = tk.Button(root, text="Plot Forecast", command=on_button_click)
plot_button.pack()

# Run the Tkinter event loop
root.mainloop()
```
