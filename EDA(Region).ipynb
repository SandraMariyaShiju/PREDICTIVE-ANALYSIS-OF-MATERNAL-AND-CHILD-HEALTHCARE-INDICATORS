# HANDLING NULL VALUES

```python
# List of state names
state_names = ['Jammu & Kashmir', 'Himachal Pradesh', 'Meghalaya', 'Sikkim', 'Uttar Pradesh', 'Arunachal Pradesh',
               'Bihar', 'Nagaland', 'West Bengal', 'Andhra Pradesh', 'A & N Islands', 'Assam', 'Mizoram',
               'Chhattisgarh', 'Chandigarh', 'Dadra & Nagar Haveli', 'Daman & Diu', 'Delhi', 'Goa', 'Gujarat',
               'Haryana', 'Jharkhand', 'Madhya Pradesh', 'Karnataka', 'Manipur', 'Kerala', 'Lakshadweep',
               'Maharashtra', 'Odisha', 'Puducherry', 'Punjab', 'Rajasthan', 'Tripura', 'Telangana', 'Tamil Nadu',
               'Uttarakhand']

# Fill null values with the mean of each state
for state in state_names:
    state_data = df[df['State'] == state]  # Select data for the current state
    state_mean = state_data.mean()  # Calculate mean values for the current state
    df.loc[df['State'] == state] = state_data.fillna(state_mean)  # Fill null values with state mean

# Now df contains null values filled with the mean of values for each state
```

```python
#checking for null values after handling null values
df.isnull().any()
```

```python
# Check for null values in each column
null_counts = df.isnull().sum()

# Iterate over columns with null values and print them
print("Columns with null values:")
for column, null_count in null_counts.iteritems():
    if null_count > 0:
        print(f"{column}: {null_count}")
```

```python
# Define the national average values
national_averages = {
    'Deliveries_Private_Institutions': 9550.53,
    'Women_Discharged_Under_48hrs_Private_Facilities': 2113.74,
    'C-Section_Deliveries_Private_Facilities': 3302.57,
    'Sterilisation_Private_Institutions': 1519.13,
    'IUCD_Insertions_Private_Facilities': 678.78,
    'Institutional deliveries at urban health facilities': 8644.70
}

# Fill null values with national averages
for column, avg_value in national_averages.items():
    df[column].fillna(avg_value, inplace=True)
```

```python
maternal_deaths_by_region_year = df.groupby(['Region', 'YearCode'])['Maternal_Deaths'].sum()

# Print the total number of maternal deaths by region and year code
print("Maternal Deaths by Region and YearCode:")
print(maternal_deaths_by_region_year)
```

```python
# Calculate the total number of maternal deaths by region and year code
maternal_deaths_by_region_year = df.groupby(['Region', 'YearCode'])['Maternal_Deaths'].sum().unstack()

# Plot the data
maternal_deaths_by_region_year.plot(kind='bar', figsize=(10, 6))
plt.title('Maternal Deaths by Region and Year')
plt.xlabel('Region')
plt.ylabel('Maternal Deaths')
plt.xticks(rotation=45)
plt.legend(title='YearCode')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Calculate the total number of reported live births and stillbirths by region and year code
live_births_by_region_year = df.groupby(['Region', 'YearCode'])['Reported_Live_Births'].sum().unstack()
still_births_by_region_year = df.groupby(['Region', 'YearCode'])['Reported_Still_Births'].sum().unstack()

# Plot the data
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 12))

live_births_by_region_year.plot(kind='bar', ax=axes[0])
axes[0].set_title('Reported Live Births by Region and Year')
axes[0].set_xlabel('Region')
axes[0].set_ylabel('Reported Live Births')
axes[0].legend(title='YearCode')
axes[0].grid(axis='y')

still_births_by_region_year.plot(kind='bar', ax=axes[1])
axes[1].set_title('Reported Still Births by Region and Year')
axes[1].set_xlabel('Region')
axes[1].set_ylabel('Reported Still Births')
axes[1].legend(title='YearCode')
axes[1].grid(axis='y')

plt.tight_layout()
plt.show()
```
