```python
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Initialize an empty DataFrame to store forecast results
forecasted_results = pd.DataFrame(columns=['State', '2020', '2021'])

# Loop through each state
for state in df1['State'].unique():
    # Subset the data for the current state
    state_data = df1[df1['State'] == state].groupby('YearCode')['Maternal_Deaths'].sum()
    
    # Train the SARIMA model for the current state
    model = SARIMAX(state_data, order=(1, 1, 1))
    sarima_results = model.fit()
    
    # Forecast for 2020 and 2021
    forecast_steps = 2  # Forecast for 2 years (2020 and 2021)
    forecast = sarima_results.forecast(steps=forecast_steps)
    
    # Round off the forecasted values to the nearest integers
    forecasted_deaths_rounded = [round(value) for value in forecast]
    
    # Append the forecasted results to the DataFrame for each state and year
    state_forecast = {
        'State': state,
        '2020': forecasted_deaths_rounded[0],
        '2021': forecasted_deaths_rounded[1]
    }
    forecasted_results = forecasted_results.append(state_forecast, ignore_index=True)

# Set 'State' column as index (optional)
forecasted_results.set_index('State', inplace=True)

# Display the forecasted results in table format
print("Forecasted number of maternal deaths (sum) in 2020 and 2021 (SARIMA) state-wise:")
print(forecasted_results)
```

```python
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Assuming your dataset has columns 'YearCode' and 'Maternal_Deaths'
# Prepare the data for SARIMA
ts = df1.set_index('YearCode')['Maternal_Deaths']

# Split the data into training (2017 and 2018) and testing (2019) sets
train_data = ts.loc[[2017, 2018]]
test_data = ts.loc[2019]

# Train the SARIMA model using data from 2017 and 2018
model = SARIMAX(train_data, order=(1, 1, 1))
sarima_results = model.fit()

# Forecast maternal deaths for 2019
forecast = sarima_results.forecast(steps=len(test_data))

# Round off the forecasted values to the nearest integers
forecast_rounded = np.round(forecast)

# Calculate MAE, MSE, and RMSE
mae = mean_absolute_error(test_data, forecast_rounded)
mse = mean_squared_error(test_data, forecast_rounded)
rmse = np.sqrt(mse)

print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
#XGBRegressor
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error

df1_model = df1.drop(columns=['State', 'District','YearCode'])

# Split data into features (X) and target (y)
X = df1_model.drop(columns=['Maternal_Deaths'])
y = df1_model['Maternal_Deaths']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Initialize XGBoost regressor
xgb_model = XGBRegressor()

# Train the model
xgb_model.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = xgb_model.predict(X_test)

from sklearn.metrics import mean_absolute_error, mean_squared_error

# Calculate MAE
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error (MAE):", mae)

# Calculate MSE
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error (MSE):", mse)

from sklearn.metrics import r2_score

# Calculate R-squared
r2 = r2_score(y_test, y_pred)
print("R-squared:", r2)

```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Initialize Random Forest Regressor
rf_model = RandomForestRegressor()

# Train the model
rf_model.fit(X_train, y_train)

# Make predictions
y_pred_rf = rf_model.predict(X_test)

# Calculate MSE and RMSE
mse_rf = mean_squared_error(y_test, y_pred_rf)

print("Random Forest:")
print("Mean Squared Error (MSE):", mse_rf)
# Calculate MAE
mae = mean_absolute_error(y_test, y_pred_rf)
print("Mean Absolute Error (MAE):", mae)

from sklearn.metrics import r2_score
# Calculate R-squared
r2_rf = r2_score(y_test, y_pred_rf)
print("R-squared (Random Forest):", r2_rf)

```

```python
##boosting methods
from sklearn.ensemble import AdaBoostRegressor

# Initialize AdaBoost Regressor
adaboost_model = AdaBoostRegressor()

# Train the model
adaboost_model.fit(X_train, y_train)

# Make predictions
y_pred_adaboost = adaboost_model.predict(X_test)

# Calculate MSE and RMSE
mse_adaboost = mean_squared_error(y_test, y_pred_adaboost)
print("AdaBoost:")
print("Mean Squared Error (MSE):", mse_adaboost)

# Calculate MAE
mae = mean_absolute_error(y_test, y_pred_adaboost)
print("Mean Absolute Error (MAE):", mae)
from sklearn.metrics import r2_score

# Calculate R-squared
r2_adaboost = r2_score(y_test, y_pred_adaboost)
print("R-squared (AdaBoost):", r2_adaboost)

```

```python
from sklearn.ensemble import GradientBoostingRegressor

# Initialize Gradient Boosting Regressor
gradientboost_model = GradientBoostingRegressor()

# Train the model
gradientboost_model.fit(X_train, y_train)

# Make predictions
y_pred_gradientboost = gradientboost_model.predict(X_test)

# Calculate MSE and RMSE
mse_gradientboost = mean_squared_error(y_test, y_pred_gradientboost)
rmse_gradientboost = mean_squared_error(y_test, y_pred_gradientboost, squared=False)

print("Gradient Boosting:")
print("Mean Squared Error (MSE):", mse_gradientboost)
#Calculate MAE
mae = mean_absolute_error(y_test, y_pred_gradientboost)
print("Mean Absolute Error (MAE):", mae)

from sklearn.metrics import r2_score

# Calculate R-squared
r2_gradientboost = r2_score(y_test, y_pred_gradientboost)
print("R-squared (Gradient Boosting):", r2_gradientboost)

```
